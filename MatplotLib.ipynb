```python
import numpy as np
from matplotlib import pyplot as plt
import pandas as pd
```

```python
x = np.arange(1,11)
```

```python
y1 = 2*x
y2 = 3*x
```

```python
x
```

```python
y1
```

```python
plt.plot(x,y1, color = 'g')
plt.plot(x,y2, color = 'y')
plt.title('two lines in a plot')
plt.xlabel('x-axis')
plt.ylabel('y-axis')

plt.show()
```

```python
# to plot multiple graphs in one screen we use "subplot"
plt.subplot(2,1,1) #indexes - 1 column 2 rows(2 subplots) and index of the subploat
plt.plot(x,y1,color = 'g')
plt.title('x vs y1')
plt.subplot(2,1,2)
plt.title('x vs y2')
plt.plot(x,y2, color = "y")
plt.xlabel('x')
plt.ylabel('y2')
plt.show()
```

```python
#bar plot
student = {"bob":87,"jp":98,"Ron":68}
names = (student.keys())
values = list(student.values())
plt.bar(names, values)
plt.title("Marks of students")
plt.xlabel("Students")
plt.ylabel("marks")
plt.grid(True)
plt.show()
```

```python
#creating a basic scatter plot
x = [4,2,5,6,8,9,7]
y = [67,56,34,89,78,54,68]
z = [90,45,24,47,88,12,23]
plt.scatter(x,y, marker = "*", c ='g', s =100)
plt.scatter(x,z, c = 'y', s = 200)
plt.show()
```

```python
#histogram plot
iris = pd.read_csv("downloads/soccer.csv")
plt.hist(iris['Overall'], color = 'r', bins = 30)
plt.show()
```

```python
#pie chart
fruits = ['apple','mangoe','guava','pineapple','kiwi']
quantity = [12,34,23,56,2]
plt.pie(quantity,labels = fruits, autopct = '%0.1f%%', colors = ['yellow','blue','black','green','grey'])
plt.show()
```
